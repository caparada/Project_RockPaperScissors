# **Rock, Paper, Scissors**

### About the game

Rock, paper, scissors is a two-player hand motion game. 

Here the player is going to play against the computer. 

To play rock, paper, and scissors, each player counts until three, shakes the fist and uses the hand to make the symbol of a rock, paper or scissors.

- **rock**: represented by a closed fist
- **paper**: represented by a open hand (palm)
- **scissors**: represented by 2 fingers

![rock, paper, and scissors hand symbols](https://www.shutterstock.com/image-vector/rock-scissors-paper-hand-gesture-260nw-689530327.jpg)

Rules of the game:
- **Rock** (fist) smashes scissors (two fingers) - Rock wins!
- **Paper** (palm) wraps rock (fist) - Paper wins!
- **Scissors** (two fingers) cut paper (palm) - Scissors win!

<font color='green'>*Note:
This is a 3-round game.

The game will display the winner after each round. At the end of all 3 rounds, the game will display the final winner.

The player will then have the chance to play again or quit the game.*</font>

Game command-line summary
1. Take player input
2. Randomly generate a computer's play
3. Compare results
4. Determine the winner/round = 1 point 
5. Count the total score to determine the winner of all 3 rounds
6. Congratulates the winner
7. Ask if the player want to keep playing or quit.

### Game set up

```python
# Load the Python module 'random'. 
# This module generates choices and numbers randomly.
# The random choices generated by the module ramdom,
# will be the computers choices generated to the player.
import random

# Defining the function get_choices().
def get_choices():
      """to get player and computer inputs. 
      It also checks if the player input is correct."""
      # Make a list to specify all choices of the game and store it in the variable throw.
      throw = ["rock", "paper", "scissors"]
      
      # This is a infinite loop to keep the game on
      while True:
            # # Display a message to the user to start the game
            print("\nMake your throw!")
            # Get player input and store it in the variable 'player_choice'.
            player_choice = input("rock, paper or scissors? ").lower()
            # Get the computer's random input based on the list 'CHOICES'
            if player_choice in throw:
                  computer_choice = random.choice(throw)
                  # Store inputs in a dictionary names CHOICES
                  CHOICES = {"player": player_choice, "computer": computer_choice}
                  return CHOICES # return results
            # Makes sure the player input is correct.
            else:
                  print(f"You typed {player_choice}. Invalid choice!\n"
                        + "Enter rock, paper, or scissors.")
               
```

```python
# Testing the function get_choices()
get_choices()
```

```python
# Defining the function win_checker().
def win_checker(player_choice, computer_choice):
    """This function determines the winner based on the game rules"""
    print(f"You chose: {player_choice}. Computer chose: {computer_choice}.")
    # if-elif-else statements were used to determine the conditions to win the game.
    if player_choice == computer_choice:
        print("It's a tie!") 
        return "none"
    
    elif player_choice == "rock" and computer_choice == "scissors":
        print("Rock smashes scissors! You win!")
        return "player"
              
    elif player_choice == "paper" and computer_choice == "rock":
        print("Paper covers rock! You win!")
        return "player"
        
    elif player_choice == "scissors" and computer_choice == "paper":
        print("Scissors shreds paper! You win!")
        return "player"
    
    else:
        print("Computer wins!")
        return "computer"
    
```

```python
# Testing the function win_checker()
player_choice = "rock"
computer_choice = "rock"

win_checker(player_choice, computer_choice)
```

### Gameplay loop & Game End:

```python
# Flagging that the game is active
active = True

# Infinite loop to keep the game running unless the player decides otherwise.
while active:
    # Counting the rounds and scores
    round = 0
    player_score = 0
    computer_score = 0
    
    # Printing the game rules to the player
    print("Rules: \nrock vs paper: paper wins\n"
    +"\nrock vs scissors: rock wins\n"
    + "\npaper vs scissors: scissors wins\n")
    
    # Determine 3 rounds of the game with the for loop.
    for n in range(3):
        # Getting player and computer inputs and storing them in the variable choices.
        choices = get_choices()
        # Getting the winners and storing the information in the variable result.
        result = win_checker(choices["player"], choices["computer"])
        # Adding 1 point to the winner's score.
        if result == 'player':
            player_score += 1
        elif result == "computer":
            computer_score += 1
        else:
            player_score == computer_score
        round += 1
        print("==================================")
    
    # GAME END:
    # Printing the total score at the end of 3 rounds.    
    print("COMPUTER SCORE: "+str(computer_score))
    print("YOUR SCORE: "+str(player_score))
    print("==================================")

    # Displaying congratulatory or loss messages to the player
    if player_score > computer_score:
        print("Congratulations!!!! You are the winner!!!!")
    elif player_score < computer_score:
        print("Try again!")
    elif player_score == computer_score:
        print("It is a tie!")
    
    # Inviting the player to play again or quit the game
    play_again = input("Wanna play again? (y/n): ")
    if play_again.lower() == "n":
        print("Goodbye.")
        break # Stops the infinite loop when the player no longer wants to play
        
```
